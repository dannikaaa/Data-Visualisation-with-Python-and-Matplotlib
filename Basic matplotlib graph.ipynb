```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[4,5,6,3])
plt.show()
```

```python
x = [1,2,3,4,5]
y = [4,7,4,7,3]
plt.plot(x,y)
plt.xlabel('Plot Number')
plt.ylabel('Random Number')
plt.title('Epic Graph tutorial for data viz in Python with Matplotlib.\nTutorial showing labels and titles')
plt.show()
#if we have a long long title then use '\n' to create a newline
```

```python
x = [1,2,3,4,5]
y = [4,7,4,7,3]

y2 = [9,2,4,19,3]

plt.plot(x,y, label='Initial Line')
plt.plot(x,y2, label='New Line')

plt.xlabel('Plot Number')
plt.ylabel('Random Number')
plt.title('Epic Graph tutorial for data viz in Python with Matplotlib.\nTutorial showing labels and titles')

plt.legend() #reveals the legend 
plt.show()
```

```python
x = [2,4,8,9]
x2 = [1,3,5,7]
y = [1,8,9,12]
y2 = [9,2,4,10]

plt.bar(x,y,label="One")
plt.bar(x2,y2,label="Two")

plt.xlabel('bar number')
plt.ylabel('bar height')
plt.title('bar chart tutorial')
plt.legend() #reveals the legend 
plt.show()
```
